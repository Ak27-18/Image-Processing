# Arithmetic & Logical Expressions.

Image arithmetic applies to the standard arithmetic operation or a logical operator of two or more images. The operators are applied in a pixel by pixel way that is the value of a pixel in the output image depends on the value of the corresponding pixels in the input images.HENCE THE IMAGES MUST BE OF SAME SIZE.

AIM:To perform arithmetic operations on image

```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

# Read & Display Image. 

```python
imgm = cv2.imread("m.jpg")
imge = cv2.imread("e.jpg")
img1 = cv2.cvtColor(imgm,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(imge,cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2);
```

# Addition

```python
add = cv2.add(img1,img2)
plt.imshow(add)
plt.title("Added Img: img1+img2");
```

# Subtraction

```python
sub = cv2.subtract(img1,img2)
plt.imshow(sub)
plt.title("subtracted Img: img1-img2");
```

# Multiplication

```python
mul = cv2.multiply(img1,img2)
plt.imshow(mul)
plt.title("multiplied Img: img1*img2");
```

# Division

```python
div = cv2.divide(img1,img2)
plt.imshow(div)
plt.title("divided Img: img1/img2");
```

# NOT

```python
bit_not = cv2.bitwise_not(img1)
plt.subplot(1,2,1)
plt.imshow(bit_not)
plt.title("NOT Img:NOT img1");
bit_not = cv2.bitwise_not(img2)
plt.subplot(1,2,2)
plt.imshow(bit_not)
plt.title("NOT Img: NOT img2");
```

# AND

```python
bit_and = cv2.bitwise_and(img1,img2)
plt.imshow(bit_and)
plt.title("And Img: img1 AND img2");
```

# OR

```python
bit_or = cv2.bitwise_or(img1,img2)
plt.imshow(bit_or)
plt.title("or Img: img1 OR img2");
```

# XOR

```python
bit_xor = cv2.bitwise_xor(img1,img2)
plt.imshow(bit_xor)
plt.title("xor Img: img1 XOR img2");
```
