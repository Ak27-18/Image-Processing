# Morphological Operations

AIM: to perform the following Morphological Operations
        
        a. Erosion
        
        b. Dilation
        
        c. Opening
        
        d. Closing
        
        e.Gradient


Morphological Operations is broad set of image processing operations that process digital images based on their shapes. In the morphological operation each pixel is corresponding to the value of the other pixel into neighborhood by choosing the shape and size of the neighborhood pixel you can construct a morphological operation that is sensitive to specifics shapes in the input image.

Erosion removes pixels on object boundaries.

Dilation adds pixels on the object boundaries.

Opening operation erodes an image and then directs the eroded image.

Closing operation directs an image and then erodes the dilated image.

Gradient operation is equal to the difference between dilation and erosion of an image.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import skimage.io as sk
import skimage.morphology as morph
from skimage import color, data, restoration
from scipy.signal import convolve2d as conv2
```

```python
#Importing the image 
img = cv2.imread("nat.jpg");
I = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(I);  
plt.title("Original image");  
```

# Erosion 

```python
#erosion
erosion=morph.erosion(I)
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,5),sharex=True,sharey=True)

ax[0].imshow(I)
ax[0].set_title('Original')
ax[1].imshow(erosion)
ax[1].set_title('Eroded')
;
```

# Dilation

```python
#dilated
dilation=morph.dilation(I)
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,5),sharex=True,sharey=True)

ax[0].imshow(I)
ax[0].set_title('Original')
ax[1].imshow(dilation)
ax[1].set_title('Dilated')
;
```

# Opening

```python
#opening
opening=morph.opening(I)
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,5),sharex=True,sharey=True)

ax[0].imshow(I)
ax[0].set_title('Original')
ax[1].imshow(opening)
ax[1].set_title('Opened')
;
```

# Closing

```python
#closing
closing=morph.closing(I)
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,5),sharex=True,sharey=True)

ax[0].imshow(I)
ax[0].set_title('Original')
ax[1].imshow(closing)
ax[1].set_title('closed')
;
```

# Gradient

```python
dilation=morph.dilation(I)
erosion=morph.erosion(I)
gradient=dilation-erosion

fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,5),sharex=True,sharey=True)

ax[0].imshow(I)
ax[0].set_title('Original')
ax[1].imshow(gradient)
ax[1].set_title('gradient')
;
```
