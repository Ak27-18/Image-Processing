# Basic Geometric Transformation

Geometric transformation is an essential image processing technique that have wide applications, a simple use would be in a computer graphics to simply re-screen the graphic content when displaying it on the desktop versus mobile it could also be applied to projectively wrap an image to another image plane

AIM: Perform the following operations on the image
        
        a.Scaling
            i.  Upscaling
            ii.	Downscaling
        b.Thresholding
        c.Translation
        d.Rotation
        e.Shear


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cat.jpg')
plt.title('Original Image')
plt.imshow(img);
```

# Scaling

```python
img = cv2.imread('cat.jpg')
scale1=cv2.resize(img,None,fx=6,fy=6,interpolation=cv2.INTER_CUBIC)
scale2=cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_AREA)

plt.subplot(1,2,1)
plt.imshow(scale1)
plt.title('Upscale Image')

plt.subplot(1,2,2)
plt.imshow(scale2)
plt.title('Downscale Image');
```

# Thresholding

```python
ret,thresh1 = cv2.threshold(img,50,250,cv2.THRESH_BINARY)
plt.imshow(thresh1,'gray');
```

# Translation

```python
img = cv2.imread("cat.jpg")
print(img.shape)
r,c,h=img.shape
```

```python
matrix=np.float32([[1,0,100],[0,1,50]])
img1=cv2.warpAffine(img,matrix,(r,c))
plt.subplot(1,2,1)
plt.imshow(img1)

matrix=np.float32([[1,0,130],[0,1,270]])
img2=cv2.warpAffine(img,matrix,(500,494))
plt.subplot(1,2,2)
plt.imshow(img2);
```

# Rotation

```python
img = cv2.imread('cat.jpg')
plt.subplot(2,2,1)
plt.imshow(img)

img1=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
plt.subplot(2,2,2)
plt.imshow(img1)

img2=cv2.rotate(img,cv2.ROTATE_180)
plt.subplot(2,2,3)
plt.imshow(img2)

img3=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.subplot(2,2,4)
plt.imshow(img3);
```

# Shear

```python
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(img)
r,c,h=img.shape

mat1=np.float32([[1,0.5,0],[0,1,0],[0,0,1]])
img1=cv2.warpPerspective(img,mat1,(int(c*1.5),int(r*1.5)))
plt.subplot(1,3,2)
plt.imshow(img1)
plt.title('Horizontal Shear')

mat2=np.float32([[1,0,0],[0.5,1,0],[0,0,1]])
img2=cv2.warpPerspective(img,mat2,(int(c*1.5),int(r*1.5)))
plt.subplot(1,3,3)
plt.imshow(img2)
plt.title('Vertical Shear');
```
