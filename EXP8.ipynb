# Edge Detection

Edge detection allows user to observe the features of an image for a significant change in the gray level, this texture indicating the end of one region in the image and the beginning of another it reduces the amount of data in an image and preserve the structural property of an image.

Methods of Edge detection:

AIM: To perform edge detection using
    
    a.	Robertâ€™s filter
    b.	Prewitts filter
    c.	Sobel filter


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Roberts Edge Detection

```python
img = cv2.imread('city.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img2,cmap='gray')
x_gradient=np.array([[1,0],[0,-1]])
y_gradient=np.array([[0,1],[-1,0]])
img3=cv2.copyMakeBorder(img2,1,1,1,1,cv2.BORDER_CONSTANT,0)
Robert_img1=cv2.filter2D(img3,-1,x_gradient)
Robert_img2=cv2.filter2D(img3,-1,y_gradient)
Robert_img=cv2.add(Robert_img1,Robert_img2)
plt.subplot(1,2,2)
plt.imshow(Robert_img,cmap='gray');
```

# Prewitts Edge Detection

```python
img = cv2.imread('city.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img2,cmap='gray')
x_gradient=np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
y_gradient=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img3=cv2.copyMakeBorder(img2,1,1,1,1,cv2.BORDER_CONSTANT,0)
Prewitt_img1=cv2.filter2D(img3,-1,x_gradient)
Prewitt_img2=cv2.filter2D(img3,-1,y_gradient)
Prewitt_img=cv2.add(Prewitt_img1,Prewitt_img2)
plt.subplot(1,2,2)
plt.imshow(Prewitt_img,cmap='gray');
```

# Sobels Edge Detection

```python
img = cv2.imread('city.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img2,cmap='gray')
x_gradient=np.array([[-1,0,-1],[-2,0,2],[-1,0,1]])
y_gradient=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
img3=cv2.copyMakeBorder(img2,1,1,1,1,cv2.BORDER_CONSTANT,0)
Sobel_img1=cv2.filter2D(img3,-1,x_gradient)
Sobel_img2=cv2.filter2D(img3,-1,y_gradient)
Sobel_img=cv2.add(Sobel_img1,Sobel_img2)
plt.subplot(1,2,2)
plt.imshow(Sobel_img,cmap='gray');
```
