# Noise Models

Noise is always present in digital image processing during image acquisition coding transmission and processing steps

Noise is A random signal it is used to destroy most part of the image information image distortion is the most pleasance problemImage processing image is distorted due to various types of noises such as Gaussian noise,Rayleigh Noise,Salt n Pepper Noise,etc.

Noise tells unwanted information in digital image noise produces undesirable effects such as artifact unrealistic edges unseen lines corners etc. To reduce this undesirable effects,Nice model is essential for further processing

AIM: To implement the various noise models on the desired image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import warnings
```

```python
img = cv2.imread('sun.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img2,cmap='gray')
plt.subplot(1,2,2)
plt.hist(img2.ravel(),256,[0,256]);plt.show()
```

# Gaussian Noise

```python
img = cv2.imread('sun.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def noisy(image):
    r,c=img2.shape
    mean=50
    var=1
    sigma=var**0.5
    gauss=np.random.normal(mean,sigma,(r,c))
    noisy=image+gauss
    return noisy

G= noisy(img2)
plt.subplot(1,2,1)
plt.imshow(G,cmap='gray')
plt.subplot(1,2,2)
plt.hist(G.ravel(),256,[0,256]);plt.show()
```

# Rayleigh Noise

```python
img = cv2.imread('sun.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def noisy(image):
    r,c=img2.shape
    mean=50
    var=1
    sigma=var**0.5
    ray=np.random.rayleigh(scale=1,size=(r,c))
    noisy=image+ray
    return noisy

R=noisy(img2)
plt.subplot(1,2,1)
plt.imshow(R,cmap='gray')
plt.subplot(1,2,2)
plt.hist(R.ravel(),256,[0,256]);plt.show()
```

# Erlang Noise

```python
img = cv2.imread('sun.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def noisy(image):
    r,c=img2.shape
    mean=50
    var=1
    sigma=var**0.5
    erlang=np.random.gamma(shape=5,scale=1,size=(r,c))
    noisy=image+erlang
    return noisy

E= noisy(img2)
plt.subplot(1,2,1)
plt.imshow(E,cmap='gray')
plt.subplot(1,2,2)
plt.hist(E.ravel(),256,[0,256]);plt.show()
```

# Exponential Noise

```python
img = cv2.imread('sun.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def noisy(image):
    r,c=img2.shape
    mean=50
    var=1
    sigma=var**0.5
    expo=np.random.exponential(scale=1,size=(r,c))
    noisy=image+expo
    return noisy

Exp= noisy(img2)
plt.subplot(1,2,1)
plt.imshow(Exp,cmap='gray')
plt.subplot(1,2,2)
plt.hist(Exp.ravel(),256,[0,256]);plt.show()
```

# Uniform Noise

```python
img = cv2.imread('sun.jpg')
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def noisy(image):
    r,c=img2.shape
    mean=50
    var=1
    sigma=var**0.5
    uni=np.random.uniform(size=(r,c))
    noisy=image+uni
    return noisy

U= noisy(img2)
plt.subplot(1,2,1)
plt.imshow(U,cmap='gray')
plt.subplot(1,2,2)
plt.hist(U.ravel(),256,[0,256]);plt.show()
```
