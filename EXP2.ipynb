# Reading images in different colour modes

Human retina happens to have a rod shaped cells that are sensitive to all lights and coned shaped cells that come in three kinds:
                
                a. Red sensitive.
                
                b. Green sensitive.
                
                c. Blue sensitive.
                
Defort there are three (additive) primary colors: red, green, blue. Color monitors and TV sets use RGB to display all the colors so every color is some amount of red some amount of green and some amount of blue

AIM:To read and display the image in various color component and combination of components. Also to convert color image to grayscale and black and white.

```python
import numpy 
import cv2 
import matplotlib.pyplot as plt
```

# Read and Display image

```python
img = cv2.imread("c.JPG")
img1 =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title('Color Image')
plt.imshow(img1);
print("shape =",img1.shape)
print("size =",img1.size)
```

# Read Color Image & Display details

```python
img = cv2.imread("c.JPG")
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print("shape =",img1.shape)
print("size =",img1.size)
img_arr = numpy.asarray(img1)
print(img_arr)
```

# Read and Display Grayscale  Image

```python
img = cv2.imread("c.JPG")
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img2 , 'gray')
plt.title("Gray Scale Image");
print("shape =",img2.shape)
print('size =',img2.size)
```

# Read Grayscale Image & Display details

```python
img = cv2.imread("c.JPG")
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print("shape =",img2.shape)
print("size =",img2.size)
img_arr = numpy.asarray(img2)
print(img_arr)
```

# Read Grayscale Image & Convert to B&W

```python
img = cv2.imread("c.JPG")
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh1 = cv2.threshold(img2,127,254,cv2.THRESH_BINARY)
plt.imshow(thresh1,'gray');
```

# Read B&W Image & Display details

```python
img = cv2.imread("c.JPG")
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh1 = cv2.threshold(img2,127,254,cv2.THRESH_BINARY)
print("size =",thresh1.size)
img_arr = numpy.asarray(thresh1)
print(img_arr)
```

# Red component

```python
r,c,p=img1.shape
temp=numpy.zeros((r,c,p),numpy.uint8)
temp[:,:,0]=img[:,:,0]
plt.imshow(temp);
```

# Green component

```python
temp1=numpy.zeros((r,c,p),numpy.uint8)
temp1[:,:,1]=img[:,:,1]
plt.imshow(temp1);
```

# Blue component

```python
temp2=numpy.zeros((r,c,p),numpy.uint8)
temp2[:,:,2]=img[:,:,2]
plt.imshow(temp2);
```

# Green-Blue component

```python
print("Green-blue")
temp2=numpy.zeros((r,c,p),numpy.uint8)
temp2[:,:,1]=img[:,:,1]
temp2[:,:,2]=img[:,:,2]
plt.imshow(temp2);
```

```python
print("red-blue")
temp2=numpy.zeros((r,c,p),numpy.uint8)
temp2[:,:,0]=img[:,:,0]
temp2[:,:,2]=img[:,:,2]
plt.imshow(temp2);
```

```python
print("red-green")
temp2=numpy.zeros((r,c,p),numpy.uint8)
temp2[:,:,0]=img[:,:,0]
temp2[:,:,1]=img[:,:,1]
plt.imshow(temp2);
```
